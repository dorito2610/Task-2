### Indexing the dataframe

```python
dataframe=pd.read_csv("colors.csv",names=["color","name","HEX value","R","G","B"],header=None)
dataframe
```

### Function to identify the colors

```python
# Global variables
x=y=r=g=b=0
mouse_click=False
```

##### We identify the colors by finding the matching color from the resultant array of RGB values

```python
def color_name(R,G,B):
    minimum=10000 # treshhold value
    for i in range(len(dataframe)):
        rgb=abs(R - int(dataframe.loc[i,"R"]))+abs(G - int(dataframe.loc[i,"G"]))+abs(B - int(dataframe.loc[i,"B"]))
        if rgb<minimum:
            minimum=rgb
            color_name=dataframe.loc[i,"name"]
    return color_name
```

### function to get the co-ordinates of the mouse click

```python
def click(lclick,X,Y,flags,param):
    if lclick==cv2.EVENT_FLAG_LBUTTON: # single left click
        # opencv uses b,g,r format instead of rgb

        global b,g,r,x,y,mouse_click
        mouse_click=True
        x=X
        y=Y
        b,g,r=image[y,x]
        b=int(b)
        g=int(g)
        r=int(r)
        
```

#### Creating a window to display the input image and responding to the mouse click

```python
cv2.namedWindow("image")
cv2.setMouseCallback("image",click)
```

```python
while True:
    cv2.imshow("image",image)
    if mouse_click==True:
        cv2.rectangle(image,(20,20),(750,60),(255,255,255),-1)
        output=color_name(r,g,b)+"RGB values="+str(r)+" "+str(g)+" "+str(b)
        cv2.putText(image,output,(50,50),2,0.8,(0,0,0),2,cv2.LINE_AA)
    mouse_click=False

    #breaking loop when user presses esc key
    if cv2.waitKey(20) & 0xFF==27:
        break
cv2.destroyAllWindows()
```
